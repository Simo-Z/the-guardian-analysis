### Create DataFrame function


```python
def get_df(url: str, params: dict):
        all_results = []
        current_page = 1
        total_pages = 1
        while current_page <= total_pages:
                params["page"] = current_page
                try:
                        r = requests.get(url, params)
                        all_results = all_results + r.json()["response"]["results"]
                        r.raise_for_status()
                except Exception as err:
                        SystemExit(err)
                if current_page == 1:
                        total_pages = r.json()['response']['pages']
                        print("---- API STATUS ---- ")
                        print("URL: ", r.url)
                        print("status",  r.json()["response"]["status"])
                        print("total",  r.json()["response"]["total"])
                        print("startIndex",  r.json()["response"]["startIndex"])
                        print("pageSize",  r.json()["response"]["pageSize"])
                        print("pages",  r.json()["response"]["pages"])
                        print("orderBy",  r.json()["response"]["orderBy"])
                        print("---- RUNTIME STATUS ---- ")

                print(f"Page ({current_page}/{total_pages}) extracted!")
                        
                current_page += 1
        return pd.DataFrame(all_results).set_index("webPublicationDate")
```

### Set Params

```python
all_content_params = {
    "page-size": "50",
    "show-fields": ["all"],
    "show-tags": ["all"],
    "show-section": "true",
    "show-blocks": ["all"],
    "show-references": ["all"],
    "show-rights": ["all"],
    "from-date": from_date,
    "api-key": os.getenv("GUARDIAN_API_KEY"),
}
```

### API Request + create DF

```python
all_articles_df = get_df(content_url, all_content_params)
all_articles_df.info()
```

### Create CSV

```python
all_articles_df.to_csv(f"../../data/raw/{file_name}_raw.csv.zip", index=True, compression="zip")
```
