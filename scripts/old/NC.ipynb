```python
url = "https://api.newscatcherapi.com/v2/search"
headers = {
    "x-api-key": os.getenv("NEWS_CATCHER_API_KEY")
}
NC_params = {
    "q":"*",
    "lang":"en",
    "sort_by":"relevancy",
    'page_size': 100,
}
```

```python
def get_df(url: str, params: dict, headers: dict):
        all_results = []
        page = 1
        total_pages = 1
        while page <= 2:
                params["page"] = page
                try:
                        r = requests.get(url, headers=headers, params=params)
                        all_results = all_results + r.json()["articles"]
                        r.raise_for_status()
                except Exception as err:
                        SystemExit(err)
                if page == 1:
                        total_pages = r.json()["total_pages"]
                        print("---- API STATUS ---- ")
                        print("status",  r.json()["status"])
                        print("total_hits",  r.json()["total_hits"])
                        print("page",  r.json()["page"])
                        print("total_pages",  r.json()["total_pages"])
                        print("page_size",  r.json()["page_size"])
                        print("---- RUNTIME STATUS ---- ")
                        
                
                print(f"Page ({page}/{total_pages}) extracted!")
                page += 1
        return pd.DataFrame(all_results).set_index("published_date")
```
