### CLIMATE CHANGE LIVE

Load API key from .env file

```python
load_dotenv()
api_key =  os.getenv("API_Key_CCLive")
```

```python
url = "https://climate-change-live243.p.rapidapi.com/news"

headers = {
	"X-RapidAPI-Host": "climate-change-live243.p.rapidapi.com",
	"X-RapidAPI-Key": api_key
}

response = requests.request("GET", url, headers=headers)

response.text
```

```python
response.json()
```

```python
df = pd.DataFrame(response.json())
df
```

```python
df.to_csv("Test.csv", index=False)
```
