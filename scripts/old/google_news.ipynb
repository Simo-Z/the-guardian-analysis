# Google API

## Import Packages

```python
import pandas as pd
import requests
import os
from dotenv import load_dotenv
```

## Load API Key from .env File

```python
load_dotenv()
api_key = os.getenv("GOOGLE_API_KEY")
```

```python
search_keyword = "climate"
start_date = "2022-04-23"
sort_by = "publishedAt"
search_url =f"https://newsapi.org/v2/everything?q={search_keyword}&from={start_date}&sortBy={sort_by}&apiKey={api_key}"
```

```python
search_result = requests.get(search_url)
```

```python
search_result.json()
 
```

```python
df = pd.DataFrame(search_result.json()["articles"])   
    
```
